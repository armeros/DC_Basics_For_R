## **Arithmetic with strings**

* Unlike R, Python allows you to use certain mathematical operators on non-numeric values. 
* When you use `+` on strings, it will concatenate them together:

```
print('hello' + 'world')

'helloworld'
```
* If you place two strings next to each other, they will also be concatenated.

```
print('hello' 'world')

'helloworld'
```

* When you multiply a string by an integer, `n`, the string will be repeated n times:

```
print('hello' * 3)

'hellohellohello'
```







````python
# Add 'the quick' to 'brown fox'
print('the quick' + 'brown fox')

# Assign 'jump' to the variable x
x = 'jump'

# Multiply x by 3
print(x * 3)

# Have the string 'lazy' next to the string 'dog'
print("lazy" "dog")
````

# **Mapping functions**

* `map()` takes the name of the function as its first argument
* then a list of values as the second argument

The specified function is then applied to all the values in the second argument one-at-a-time.

````python
def binge_male(num_drinks):
    if num_drinks <= 5:
        return 'non-binge'
    else:
        return 'binge'

def binge_female(num_drinks):
    if num_drinks <= 4:
        return 'non-binge'
    else:
        return 'binge'

num_drinks = [5, 4, 3, 3, 3, 5, 6, 10]

# map the binge_male function to num_drinks
print(list(map(binge_male, num_drinks)))

# map the binge_female function to num_drinks
print(list(map(binge_female, num_drinks)))
````

## **Converting string to categorical variable**

You can convert a column into a categorical data type by passing '`category`' to the `.astype()` method

* Use `.astype("category")`
* Use `.cat.categories` to see the unique values of categories
* Use `.cat.codes` to see the numeric values for each category

Another use case for categorical values is when you want to preserve ordering in your data. 

* `df['column_name'].cat.reorder_categories(['low', 'high'], ordered=True)`

### **Category**

* Convert the type of 'time' column into category and print the categories in this column.
* Order the time category so `'Lunch'` is before `'Dinner'` and assign it to the `'time2'` column and print the categories in this column.


````python
import pandas as pd

tips = pd.read_csv("tips.csv")

# Convert the type of time column
tips['time'] = tips['time'].astype('category')

# Use the cat accessor to print the categories in the time column
print(tips['time'].cat.categories)

# Order the time category so lunch is before dinner
tips['time2'] = tips["time"].cat.reorder_categories(["Lunch", "Dinner"], ordered=True)

# Use the cat accessor to print the categories in the time2 column
print(tips["time2"].cat.categories)
````

## **Datetime in Pandas**

* Use the `pd.to_datetime()` function from pandas to convert a column with strings of datetime into datetime object
 * You need to specify the format argument that specifies in which format the dates exist, e.g., If the dates are in the month/day/year format, you would use:
 `pd.to_datetime(df['date_column_as_string'], format='%m/%d/%Y')`
* Use `.dt.year`, `.dt.day`, .dt.month to access datetime component.

### **Dates (I)**
* Import the `country_timeseries.csv` dataset and assign it to ebola and inspect the type of `'Date'` column.

* Convert the type of `'Date'` column into datetime using the given format and inspect the type of `'Date'` column.

### ** Dates {II)**

You can import the data while parsing the dates correctly.

* Import the dataset and ensure that the `'Date'` column is imported as `datetime` type. After importing, inspect the type of 'Date' column.

* Extract the year, month and day from the `'Date'` column and then inspect these newly created columns.

````python
import pandas as pd

# Load the country_timeseries dataset
ebola = pd.read_csv('country_timeseries.csv')

# Inspect the Date column
print(ebola['Date'].dtype)

# Convert the type of Date column into datetime
ebola['Date'] = pd.to_datetime(ebola["Date"], format='%m/%d/%Y')

# Inspect the Date column
print(ebola['Date'].dtype)

# Load the dataset and ensure Date column is imported as datetime
ebola = pd.read_csv('country_timeseries.csv', parse_dates=["Date"])

# Inspect the Date column
print(ebola['Date'].dtype)

# Create a year, month, day column using the dt accessor
ebola['year'] = ebola.Date.dt.year
ebola['month'] = ebola.Date.dt.month
ebola['day'] = ebola.Date.dt.day

# Inspect the newly created columns
print(ebola[['year', 'month', 'day']].head())
````

# **Missing values**

* `pd.isnull(df['column'])` will return True if the value is missing, or False if there are no missing values.

* You can also recode missing values with the `.fillna()` method.

````python
import pandas as pd

tips = pd.read_csv("tips.csv")
# Print the rows where total_bill is missing
print(tips.loc[pd.isnull(tips['total_bill'])])
print(tips.isnull().sum())

# Mean of the total_bill column
tbill_mean = tips['total_bill'].mean()

# Fill in missing total_bill
print(tips['total_bill'].fillna(tbill_mean))
````

# **Facet plots in seaborn**

* Create a scatter plot of `'total_bill'` on the x-axis and `'tip'` on the y-axis with `lmplot()`.
* Facet the plot by `'smoker'` and color the points by `'sex'`.
* Use the `FacetGrid()` function to create a facet object facetted by `'time'` and `'smoker'`, and colored by `'sex'`.
* Call .`map()` on facet to generate a facetted scatter plot of `'total_bill'` and `'tip'`.

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

tips = pd.read_csv("tips.csv")

# Scatter plot of total_bill and tip faceted by smoker and colored by sex
sns.lmplot(x='total_bill', y='tip', data=tips, hue="sex", col="smoker")
plt.show()

# FacetGrid of time and smoker colored by sex
facet = sns.FacetGrid(tips, col="time", row="smoker", hue="sex")

# Map the scatter plot of total_bill and tip to the FacetGrid
facet.map(plt.scatter, "total_bill", "tip")
plt.show()
````

# **Load multiple data files**

* When you want to import a bunch of datasets without having to make multiple read_csv() calls
* You can use the `glob` library
* The `glob()` function returns a list of filenames that match a specified pattern.

````python
import glob
import pandas as pd
import matplotlib.pyplot as plt

# Get a list of all the csv files
csv_files = glob.glob('*.csv')

# List comprehension that loads of all the files
dfs = [pd.read_csv(df) for df in csv_files]

# List comprehension that looks at the shape of all DataFrames
print([df.shape for df in dfs])

# Get the planes DataFrame
planes = dfs[0]

# Count the frequency of engines in our data
print(planes['engines'].value_counts())

# Look at all planes with >= 3 engines
print(planes.loc[planes['engines'] >= 3])

# Look at all planes with >= 3 engines and <= 100 seats
print(planes[(planes["engines"] >= 3) & (planes["seats"] <= 100)])

# Scatter plot of engines and seats
planes.plot(x='engines', y='seats', kind="scatter")
plt.show()
plt.tight_layout()
# Histogram of seats
planes["seats"].plot(kind="hist")
plt.show()
plt.tight_layout()
# Boxplot of seats by engine
planes.boxplot(column="seats", by="engine")
plt.xticks(rotation=45)
plt.show()
plt.tight_layout()
````

````python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns


def get_season(time_hour):
    y_m_dt = time_hour.split('-')
    month = int(y_m_dt[1])
    
    d_t = y_m_dt[2].split(' ')
    day = int(d_t[0])
    return(calculate_season(month, day))

def calculate_season(month, day):
    if month < 3:
        return('winter')
    elif month == 3:
        if day < 20:
            return('winter')
        else:
            return('spring')
    elif month < 6:
        return('spring')
    elif month == 6:
        if day < 21:
            return('spring')
        else:
            return('summer')
    elif month < 9:
        return('summer')
    elif month == 9:
        if day < 22:
            return('summer')
        else:
            return('fall')
    elif month < 12:
        return('fall')
    elif month == 12:
        if day < 21:
            return('fall')
        else:
            return('winter')
    else:
        return(np.NaN)

flights = pd.read_csv("flights.csv")
print(flights["time_hour"])

# Apply the function on data
flights['season'] = flights['time_hour'].apply(get_season)

# Print time_hour and season
print(flights[["time_hour", "season"]])

# Calculate total_delay
flights['total_delay'] = flights['dep_delay'] + flights['arr_delay']

# Mean total_delay by carrier
tdel_car = flights.groupby('carrier')['total_delay'].mean().reset_index()
print(tdel_car)

# Mean dep_delay and arr_delay for each season
dadel_season = flights.groupby('season')['dep_delay', 'arr_delay'].mean().reset_index()
print(dadel_season)

# Mean and std delays by origin
del_ori = flights.groupby('origin')['total_delay', 'dep_delay', 'arr_delay'].agg(["mean", 'std'])
print(del_ori)

# Create a figure
fig, (ax1, ax2) = plt.subplots(2, 1)

# Boxplot and barplot in the axes
sns.boxplot(x='origin', y='dep_delay', data=flights, ax=ax1)
sns.barplot(x='carrier', y='total_delay', data=tdel_car, ax=ax2)

# Label axes
ax1.set_title('Originating airport and the departure delay')

# Use tight_layout() so the plots don't overlap
fig.tight_layout()
plt.show()

flights_sub = flights[["year",  "month",  "day",  "dep_time",  "dep_delay" , "origin"]]

# Look at the head of flights_sub
print(flights_sub.head())

# Create dummy variables
flights_dummies = pd.get_dummies(flights_sub)

# Look at the head of flights_dummies
print(flights_dummies.head())
````
